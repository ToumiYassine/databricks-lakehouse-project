```sql

INSERT OVERWRITE lakehouse.gold.dim_products (
  product_nk,
  prd_key,
  prd_name,
  prd_cost,
  prd_start_dt,
  prd_end_dt,
  prd_key_sales,
  prd_line,
  category_id,
  category,
  subcategory,
  maintenance,
  load_ts
)
SELECT
  p.prd_id            AS product_nk,
  p.prd_key,
  p.prd_name,
  CAST(p.prd_cost AS DECIMAL(18,2)) AS prd_cost,
  p.prd_start_dt,
  p.prd_end_dt,
  p.prd_key_sales,
  p.prd_line,
  p.Categoty_id       AS category_id,
  p.Category          AS category,
  p.Subcategory       AS subcategory,
  p.Maintenance       AS maintenance,
  CURRENT_TIMESTAMP() AS load_ts
FROM v_stg_products p
WHERE p.rn = 1;
```

```sql

CREATE OR REPLACE TABLE lakehouse.silver.dim_customers_1 AS
SELECT
    p.Customer_id,
    p.CUST_ID,
    p.Date_birth,
    p.customer_gender ,
    l.Country
      
FROM lakehouse.silver.erp_cust as p
LEFT JOIN lakehouse.silver.erp_location as l
ON p.CUST_ID = l.CUST_ID ;

```

```sql

CREATE OR REPLACE TABLE lakehouse.gold.dim_customers AS
SELECT
    p.cst_id,
    c1.Customer_id,
    c1.CUST_ID,
    c1.Date_birth,
    c1.customer_gender ,
    c1.Country,
    p.cst_firstname,
    p.cst_lastname,
    p.cst_marital_status,
    p.cst_gndr,
    p.cst_create_date
    
      
FROM lakehouse.silver.crm_cust_info as p
LEFT JOIN lakehouse.silver.dim_customers_1 as c1
ON p.cst_key = c1.CUST_ID ;

```

```sql

CREATE OR REPLACE TABLE lakehouse.gold.fact_sales AS
SELECT
    p.sls_ord_num,
    p.sls_prd_key,
    p.sls_cust_id,
    p.sls_order_dt,
    p.sls_ship_dt,
    p.sls_due_dt,
    p.sls_price,
    p.sls_sales,
    p.sls_quantity
      
FROM lakehouse.silver.crm_sales as p


```
