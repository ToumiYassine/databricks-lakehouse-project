```python
# 1. Lire la table Silver
crm_sales_df = spark.table("lakehouse.silver.crm_sales")

# 2. Sélection finale (équivalent SQL)
fact_sales_df = crm_sales_df.select(
    "sls_ord_num",
    "sls_prd_key",
    "sls_cust_id",
    "sls_order_dt",
    "sls_ship_dt",
    "sls_due_dt",
    "sls_sales",
    "sls_quantity",
    "sls_price"
)

# 3. Écriture Gold layer
fact_sales_df.write.mode("overwrite").format("delta").saveAsTable("lakehouse.gold.fact_saless")

print("Gold fact table fact_sales created successfully")
```
