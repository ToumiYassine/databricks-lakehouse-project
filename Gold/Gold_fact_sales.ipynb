```python
""""
from pyspark.sql.functions import col

# 1. Lire les tables Silver
cust_df = spark.table("lakehouse.silver.erp_custt")
loc_df = spark.table("lakehouse.silver.erp_locationn")

# 2. Left join
silver_dim_df = cust_df.join(
    loc_df,
    cust_df["CUST_ID"] == loc_df["CUST_ID"],
    how="left"
)

# 3. Sélection finale
silver_dim_df = silver_dim_df.select(
    col("Customer_id"),
    col("CUST_ID"),
    col("Date_birth"),
    col("Customer_Gender").alias("customer_gender"),
    col("Country")
)

# 4. Écriture Silver layer
silver_dim_df.write.mode("overwrite").format("delta").saveAsTable("lakehouse.silver.dim_customers_2")

print("Silver dimension dim_customers_1 created successfully")
```

```python
""""
from pyspark.sql.functions import col

# 1. Lire les tables Silver
crm_cust_df = spark.table("lakehouse.silver.crm_cust_info")
dim_cust_df = spark.table("lakehouse.silver.dim_customers_2")

# 2. Left join
gold_dim_cust_df = crm_cust_df.join(
    dim_cust_df,
    crm_cust_df["cst_key"] == dim_cust_df["CUST_ID"],
    how="left"
)

# 3. Sélection finale
gold_dim_cust_df = gold_dim_cust_df.select(
    col("cst_id"),
    col("Customer_id"),
    col("CUST_ID"),
    col("Date_birth"),
    col("customer_gender"),
    col("Country"),
    col("cst_firstname"),
    col("cst_lastname"),
    col("cst_marital_status"),
    col("cst_gndr"),
    col("cst_create_date")
)

# 4. Écriture Gold layer
gold_dim_cust_df.write.mode("overwrite").format("delta").saveAsTable("lakehouse.gold.dim_customerss")

print("Gold dimension dim_customers created successfully")
```
