```sql
CREATE OR REPLACE TABLE lakehouse.silver.crm_prd AS
SELECT
    CAST(prd_id AS INT) AS prd_id,
    CAST(prd_key AS STRING) AS prd_key,
    CAST(prd_cost AS DOUBLE) AS prd_cost,
    CAST(prd_name AS STRING) AS prd_name,
    CAST(prd_start_dt AS DATE) AS prd_start_dt,
    CAST(prd_end_dt AS DATE) AS prd_end_dt,
    
    CASE prd_line
        WHEN 'S' THEN 'Sport'
        WHEN 'R' THEN 'Racing'
        WHEN 'M' THEN 'Medium'
        WHEN 'T' THEN 'Turbo/Tech/Touring'
        ELSE prd_line
    END AS prd_line  -- overwrite original column

FROM lakehouse.bronze.crm_prd;
```
