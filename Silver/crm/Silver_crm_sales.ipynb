```sql
CREATE OR REPLACE TABLE lakehouse.silver.crm_sales AS
SELECT
    CAST(sls_ord_num AS STRING) AS sls_ord_num,
    CAST(sls_prd_key AS STRING) AS sls_prd_key,
    CAST(sls_cust_id AS STRING) AS sls_cust_id,
    
    -- Use TRY_TO_DATE instead of TO_DATE
    TRY_TO_DATE(CAST(sls_order_dt AS STRING), 'yyyyMMdd') AS sls_order_dt,
    TRY_TO_DATE(CAST(sls_ship_dt AS STRING), 'yyyyMMdd') AS sls_ship_dt,
    TRY_TO_DATE(CAST(sls_due_dt AS STRING), 'yyyyMMdd') AS sls_due_dt,
    
    CAST(sls_sales AS DOUBLE) AS sls_sales,
    CAST(sls_quantity AS INT) AS sls_quantity,
    CAST(sls_price AS DOUBLE) AS sls_price  

FROM lakehouse.bronze.crm_sales;
```
