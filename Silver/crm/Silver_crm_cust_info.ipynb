```sql
SELECT cst_id, COUNT(*) AS cnt
FROM lakehouse.bronze.crm_cust_info
GROUP BY cst_id
HAVING COUNT(*) > 1;
```

```sql
CREATE OR REPLACE TABLE lakehouse.bronze.crm_cust_info AS
WITH ranked AS (
  SELECT *,
         ROW_NUMBER() OVER (PARTITION BY cst_id ORDER BY cst_create_date DESC) AS rn
  FROM lakehouse.bronze.crm_cust_info
)
SELECT *
FROM ranked
WHERE rn = 1;
```

```sql
SELECT
    TRIM(cst_firstname) AS cst_firstname,
    TRIM(cst_lastname) AS cst_lastname,
    TRIM(cst_marital_status) AS cst_marital_status,
    TRIM(cst_gndr) AS cst_gndr
FROM lakehouse.bronze.crm_cust_info;
```

```sql
SELECT
    CASE cst_marital_status
        WHEN 'S' THEN 'Single'
        WHEN 'M' THEN 'Married'
        ELSE cst_marital_status
    END AS cst_marital_status,
    
    CASE cst_gndr
        WHEN 'F' THEN 'Female'
        WHEN 'M' THEN 'Male'
        ELSE cst_gndr
    END AS cst_gndr
FROM lakehouse.bronze.crm_cust_info;
```

```sql
SELECT
    INITCAP(cst_firstname) AS cst_firstname,  -- Title case: john -> John
    INITCAP(cst_lastname) AS cst_lastname
                       -- Uppercase IDs
FROM lakehouse.bronze.crm_cust_info;
```

```sql
CREATE OR REPLACE TABLE lakehouse.silver.crm_cust_info AS
SELECT
    CAST(cst_id AS INT) AS cst_id,
    CAST(cst_key AS STRING) AS cst_key,
    INITCAP(REGEXP_REPLACE(TRIM(cst_firstname), '\\s+', ' ')) AS cst_firstname,
    INITCAP(REGEXP_REPLACE(TRIM(cst_lastname), '\\s+', ' ')) AS cst_lastname,
    CASE TRIM(cst_marital_status)
        WHEN 'S' THEN 'Single'
        WHEN 'M' THEN 'Married'
        ELSE cst_marital_status
    END AS cst_marital_status,
    CASE TRIM(cst_gndr)
        WHEN 'F' THEN 'Female'
        WHEN 'M' THEN 'Male'
        ELSE cst_gndr
    END AS cst_gndr ,
    CAST(cst_create_date AS DATE) AS cst_create_date 
FROM lakehouse.bronze.crm_cust_info;
```
